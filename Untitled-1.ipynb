```python
df_2["duration"]=np.log1p(df_2["duration"])
sns.boxplot(df_2["duration"])
plt.show()
```

```python

Q1 = df_2["duration"].quantile(0.25)
Q3 = df_2["duration"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_2.loc[df_2["duration"] < lower_bound, "duration"] = lower_bound  

df_2.loc[df_2["duration"] > upper_bound, "duration"] = upper_bound

sns.boxplot(df_2["duration"])
plt.show()
```
